```python
import xarray as xr
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import numpy as np
%matplotlib inline
```

```python
dateSeries = np.array([ (datetime(2020,1,1) + timedelta(days=int(i))) for i in np.arange(100)])
da = xr.DataArray(dims=['time'],data=np.arange(100),coords={'time':dateSeries}, attrs=({'units':'random','long_name': 'random'}))
xds = xr.Dataset({'random':da})
```

```python
mask = xds['random']%6 == 0
mask = np.logical_or(mask, xds['random']%7 == 0)
mask = np.logical_or(mask, xds['random']%8 == 0)
mask = np.logical_or(mask, xds['random']%9 == 0)
var = xr.where(mask, np.nan, xds['random'])
```
