```python
from netCDF4 import Dataset
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ds = Dataset('GLORYS12v1_dailyAvg_2020-03-01_added.nc')
ssh = np.array(ds.variables['zos'])
lat = np.array(ds.variables['latitude'])
lon = np.array(ds.variables['longitude'])
depth = np.array(ds.variables['depth'])
```

```python
plt.figure(figsize=(15,5))
plt.pcolormesh(lon, lat, ssh[0,:,:], vmin=0, vmax = 0.6)
x = np.array([-103.5,-101])
y = np.array([12,13.5])
plt.scatter(x,y,s= 0.5, color = 'r')
plt.colorbar()
```

```python
xx = np.array(len(lat)* [lon])
yy = np.array(len(lon)* [lat]).T  
mask1 = np.logical_and(xx > -103.5, xx < -101)
mask2 = np.logical_and(yy > 12, yy < 14)
mask = np.logical_and(mask1, mask2)
```

```python
plt.figure(figsize=(15,5))
plt.pcolormesh(lon, lat, ssh[0,:,:], vmin=0, vmax = 0.6)
plt.scatter(x,y,s= 0.1, color = 'r')
#plt.pcolormesh(lon, lat, mask, vmin=0, vmax = 0.6, alpha=0.2)
plt.colorbar()
```

```python
lons = xx[mask]
lats = yy[mask]
```
