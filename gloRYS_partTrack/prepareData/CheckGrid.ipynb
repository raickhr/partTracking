```python
import xarray as xr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
xds = xr.open_dataset('glorysGrid.nc')
```

```python
dx = xds['dx'].to_numpy()
dy = xds['dy'].to_numpy()
area = xds['area'].to_numpy()
lat = xds['lat'].to_numpy()
lon = xds['lon'].to_numpy()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,4))
ax = axes[0]
m = ax.pcolormesh(lon, lat, dx)
plt.colorbar(m, ax=ax)

ax = axes[1]
m = ax.pcolormesh(lon, lat, dy)
plt.colorbar(m, ax=ax)

ax = axes[2]
m = ax.pcolormesh(lon, lat, area)
plt.colorbar(m, ax=ax)
```

```python
xds.close()
```
