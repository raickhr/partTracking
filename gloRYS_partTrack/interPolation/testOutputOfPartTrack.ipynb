```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset, date2num, num2date
%matplotlib inline
```

```python
xds= xr.open_dataset('/srv/data1/particleTrack_GLORYS/partTracking/gloRYS_partTrack/particleTracks.nc')
```

```python
xds
```

```python
x = xds['xpos'].to_numpy()
y = xds['ypos'].to_numpy()
z = xds['zpos'].to_numpy()

uo = xds['uo'].to_numpy()
vo = xds['vo'].to_numpy()
wo = xds['wo'].to_numpy()

p = xds['pres'].to_numpy()
dx_p = xds['dx_pres'].to_numpy()
dy_p = xds['dy_pres'].to_numpy()
rho = xds['rho'].to_numpy()
f = xds['f'].to_numpy()
xds.close()
```

```python
colors = plt.cm.jet(np.linspace(0,1,x.shape[1]))
for i in range(x.shape[1]):
    plt.plot(x[:,i],y[:,i], label = z[0,i], color=colors[i])
plt.legend()
#plt.scatter(x,y, c=p, s=1)
```

```python
D_Dt_u = (np.roll(uo,-1,axis=0) - np.roll(uo,1,axis=0))[1:-1,:]/(2*3*3600)
presXgrad = -1/rho * dx_p
xCoriolis = -f*vo
RHS = presXgrad + xCoriolis
RHS = RHS[1:-1,:]
#RHS = (np.roll(RHS,-1,axis=0) + RHS)[1:-1,:]/2

fig, axes = plt.subplots(nrows = 4, ncols=4, figsize = (20,20) )
for i in range(16):
    ii = int(i//4)
    jj = i%4
    ax = axes[ii,jj]
    ax.plot(D_Dt_u[:,i], label = 'LHS')
    ax.plot(RHS[:,i], label = 'RHS')
    text = f'particle initially \n at depth {z[0,i]} m'
    fig_x, fig_y = 0.2, 0.8
    ax.text(fig_x, fig_y, text, color ='white', horizontalalignment='center',
            weight='heavy', verticalalignment='center', transform=ax.transAxes,
            bbox=dict(facecolor='black', alpha=1))
    ax.legend()
```

```python
D_Dt_v = (np.roll(vo,-1,axis=0) - np.roll(vo,1,axis=0))[1:-1,:]/(2*3*3600)
presYgrad = -1/rho * dy_p
yCoriolis = f*uo
RHS = presYgrad + yCoriolis
RHS = RHS[1:-1,:]
#RHS = (np.roll(RHS,-1,axis=0) + RHS)[1:-1,:]/2

fig, axes = plt.subplots(nrows = 4, ncols=4, figsize = (20,20) )
for i in range(16):
    ii = int(i//4)
    jj = i%4
    ax = axes[ii,jj]
    ax.plot(D_Dt_v[:,i], label = 'LHS')
    ax.plot(RHS[:,i], label = 'RHS')
    text = f'particle initially \n at depth {z[0,i]} m'
    fig_x, fig_y = 0.2, 0.8
    ax.text(fig_x, fig_y, text, color ='white', horizontalalignment='center',
            weight='heavy', verticalalignment='center', transform=ax.transAxes,
            bbox=dict(facecolor='black', alpha=1))
    ax.legend()
```
