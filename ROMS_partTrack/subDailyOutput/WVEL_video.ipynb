```python
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np
from IPython.display import HTML
%matplotlib inline
```

```python
xds = xr.open_mfdataset('/home/shikhar.rai/WHOI/particleTrack_GLORYS/partTracking/gloRYS_partTrack/prepareData/DATA/*.nc')
```

```python
wvel = xds['wo'].to_numpy()
#xds['wo']..sel(level=level)
```

```python
lat = xds['latitude'].to_numpy()
lon = xds['longitude'].to_numpy()
```

```python
fig, ax = plt.subplots(figsize=(20,6))


m= ax.pcolormesh(lon, lat, wvel[0,0,:,:], cmap='coolwarm', vmin=-5e-5, vmax=5e-5)
plt.colorbar(m, ax=ax)

# def animate(i):
#     m.set_array(wvel[i,0,:,:].flatten())

# ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(wvel))
# ani.save('wvel.gif', writer=matplotlib.animation.PillowWriter(fps=60))
# #writer = 
# #HTML(ani.to_jshtml())
```
