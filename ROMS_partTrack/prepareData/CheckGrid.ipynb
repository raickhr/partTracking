```python
import xarray as xr
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
%matplotlib inline
```

```python
xds = xr.open_dataset('DATA_subDaily/romsGrid_WEP.nc')
dx = xds['dx'].to_numpy()
dy = xds['dy'].to_numpy()
area = xds['area'].to_numpy()
lat = xds['lat'].to_numpy()
lon = xds['lon'].to_numpy()
xds.close()
```

```python
curDateTime = datetime(2018,3,12)
xds = xr.open_dataset('ocean_his.nc')
xds = xds.sel(ocean_time=slice(f'{curDateTime.year:04d}-{curDateTime.month:02d}-{curDateTime.day:02d} {curDateTime.hour:02d}:{curDateTime.minute:02d}:{curDateTime.second:02d}',
                              f'{curDateTime.year:04d}-{curDateTime.month:02d}-{curDateTime.day:02d} {curDateTime.hour:02d}:{curDateTime.minute:02d}:{curDateTime.second:02d}'))
```

```python
pm = xds['pm']
pn = xds['pn']
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20,10))

ax = axes[0,0]
m = ax.pcolormesh(lon, lat, dx)
plt.colorbar(m, ax=ax)

ax = axes[0,1]
m = ax.pcolormesh(lon, lat, dy)# vmin = 6800, vmax = 7000)
plt.colorbar(m, ax=ax)

ax = axes[1,0]
m = ax.pcolormesh(lon, lat, 1/pm)
plt.colorbar(m, ax=ax)

ax = axes[1,1]
m = ax.pcolormesh(lon, lat, 1/pn)# vmin = 6800, vmax = 7000)
plt.colorbar(m, ax=ax)

ax = axes[2,0]
m = ax.pcolormesh(lon, lat, pn*pm)# vmin=0, vmax=3)
plt.colorbar(m, ax=ax)

ax = axes[2,1]
m = ax.pcolormesh(lon, lat, dy*pn, vmin=0, vmax=3)
plt.colorbar(m, ax=ax)
```
